# Classify movies based on rating

```python
data.columns
```

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"

    
    elif rating>=5.0:
        return "Good"

    else:
        return "Average"

data['Rating_category']=data['Rating'].apply(rating)
```

# Find unique values from genre

```python
data.columns
```

```python
data['Genre']
```

```python
list1=[]
for value in data['Genre']:
   list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```

```python
len(uni_list)
```
